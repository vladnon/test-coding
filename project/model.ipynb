```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score, recall_score
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("./exams.csv")
df
```

```python
new_df = pd.read_csv("./Student_Marks.csv")
new_df
```

```python
df.corr()
```

```python
x = df.drop("Сдача экзамена", axis=1)
y = df["Сдача экзамена"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
y_pred = model.predict(x_train)
```

```python
r2_score(y_train, y_pred)
```

```python
recall_score(y_train, y_pred)
```

```python
mean_squared_error(y_train, y_pred)
```

```python
df["сдача предсказанная"] = model.predict(x)
df["правильно/неправильно"] = df["Сдача экзамена"] == df["сдача предсказанная"]
df
df
```
